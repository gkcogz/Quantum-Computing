# Probabilistic bits

Probabilistic bit is a 2-dimensional vector $\lfloor p \rceil = \myvector{p_0 \\ p_1 }$

### Example 1

Define a random vector $\lfloor p \rceil$ and compute the probability of measuring a given value of the bit, if the value is given as a basis vector
$$
\lfloor 0 \rceil = \myvector{1 \\ 0},\\
\lfloor 1 \rceil = \myvector{0 \\ 1}.
$$

Reminder. We have seen in the previous lecture: Dot product, inner product, kroenecker product. 

We will define a function the calculates the dot product $( \lfloor p \rceil, \lfloor j \rceil ) = p_j$

We compute a new vector $w=u\otimes v$ and $L=M\otimes N$, then transform $w$ by $L$. We compare this with the tensor product $(Mu)\otimes(Nv)$.


```python
from random import randrange
from math import sqrt

# vector p is defined.
p = [0,0]

# == in range(0,101); 101 is the upper limit.

p[0] = randrange(101);
p[1] = randrange(101);

# The term "norm" is a bit misleading here, normally,
# in vector algebra, the norm of a vector is a measure of its 
# its "length" or magnitude. 

for k in range (0,2):
    print(p[k])

norm_real = sqrt(p[0]+p[1])
norm = sum(p); 

for j in range(len(p)):
        p[j] = p[j]/norm;   
        
print(p)
print(norm_real)
```

```python
# Once again, p and v are vector.
# The prerequisite: s has to be defined.  

def probability(p, value):
    s = 0
    for j in range(len(p)):
        s = s+p[j]*value[j]
    return s
```

```python
# After we defined the function, first input: vector, second input: a vector.
# Something, interesting about Jupyter Notebook, once you define a variable in one of the cells, this variable will be 
# visible to other cells. And if you implement any change to this variable, it is also taken into account.

print ("p_0 =", probability(p, [1,0]))
print ("p_1 =", probability(p, [0,1]))
```

# Correlations

We say that two p-bits are independent if we can write their state $\lfloor a \rceil$ as some product $\lfloor a \rceil = \lfloor p \rceil \otimes \lfloor q \rceil$. If we cannot, then we say that they are correlated.
