### Example 1

We define a quantum circuit that will define the operations we make. We'll use the 'StatevectorSimulator' to compute the resulting quantum state, and we'll plot the results using the helper functions.

```python
# Statevector is to debug your code.
# The default state is |0>.

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import StatevectorSimulator
from qiskit.visualization import plot_histogram
import numpy as np

# We define a quantum circuit with one qubit and one bit
q1 =  QuantumRegister(1) # Quantum register with a single qubit.
c1 = ClassicalRegister(1) # Classical register with a single bit.
qc1 = QuantumCircuit(q1,c1) # Quantum circuit with quantum and classical registers.

rotation_angle = np.pi/3

# Rotate the qubit with rotation_angle.
# We rotate the qubit around y-axis.
qc1.ry(2*rotation_angle,q1[0]) 

# Draw the circuit. 
# qubits can exist in a superposition state, where they can be both 0 and 1 simultaneously. 
# This property and entanglement and interference enable quantum circuits to perform complex 
# calculations.

display(qc1.draw(output='mpl'))

# Execute the program 1000 times.
job = StatevectorSimulator().run(qc1)

# Print the results.
statevector1 = job.result().get_statevector(qc1).data
print(statevector1)

```

### Example 2

__Quantum Gates__

_Operations on qubits are performed using quantum gates, which are the quantum analogs of classical logic gates. However, quantum gates are reversible and can operate on superpositions of qubits, enabling operations that are not possible in classical computing._


Now we will apply the $Z$ gate to check the difference between global and relative phases

adding a relative phase means we get $\myvector{a \\ -b}$

__For a relative phase only one argument differs in phase.__

adding a global phase means we get $\myvector{-a \\ -b}$

__Both arguments underwent the same phase change.__



$\ket1 = \myvector{0\\1}$ 
$\ket0 = \myvector{1\\0}$

$Z\ket1$ = $\mymatrix{cc}{ 1 & 0 \\ 0 & -1 } $  $\myvector{0\\1}$  = $ \myvector{0\\-1}$

However, in quantum computing, we typically ignore global phase factors when considering the physical state of a qubit because they do not affect measurement outcomes.

In terms of measurement outcomes, meaning the Z gate leaves the probability amplitudes of the
∣1⟩ state unchanged but inverts its phase.

its phase.

```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit

# Aer provides the tools needed to simulate quantum circuits on classical computers, allo-
# wing developers and researchers to test and validate quantum algorithms without the need for
# a quantum computer.
from qiskit_aer import StatevectorSimulator
from qiskit.visualization import plot_histogram
import numpy as np

# We define a quantum circuit with one qubit and one bit.
q2 =  QuantumRegister(1) # Quantum register with a single qubit.
c2 = ClassicalRegister(1) # Classical register with a single bit.
qc2 = QuantumCircuit(q2,c2) # quantum circuit with quantum and classical registers


# Apply Z gate.
qc2.z(q2[0])

# Draw the circuit.
display(qc2.draw(output='mpl'))


# Execute the program 1000 times.
job = StatevectorSimulator().run(qc2)

# Print the results.
statevector2 = job.result().get_statevector(qc2).data

print(statevector2)

draw_qubit()


draw_quantum_state(statevector2[0].real,statevector2[1].real,"")

# Drawing the angle with |0>-axis
show_plt()
```

Now append measurement to `qc2` and obtain measurement statistics.

```python
# Measure the first qubit, and store the result in the first classical bit.
qc2.measure(q2[0],c2[0])

# Draw the circuit by using matplotlib.
display(qc2.draw(output='mpl')) 

# Execute the circuit and read the results.
job = StatevectorSimulator().run(qc2)

counts2 = job.result().get_counts(qc2)
display(plot_histogram(counts2, title='Quantum coin flipping'))
print(counts2)
```

### Example 4

Now rotate the qubit by a random angle and apply the $Z$ gate to it.


```python
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import StatevectorSimulator
from qiskit.visualization import plot_histogram
import numpy as np

# We define a quantum circuit with one qubit and one bit.
q4 =  QuantumRegister(1) # quantum register with a single qubit
c4 = ClassicalRegister(1) # classical register with a single bit
qc4 = QuantumCircuit(q4,c4) # quantum circuit with quantum and classical registers

rotation_angle = np.pi/3
qc4.ry(2*rotation_angle,q4[0]) 

# Draw the circuit.
display(qc4.draw(output='mpl'))

qc4.measure(q4[0],c4[0])

# Execute the program 10 times.
job = StatevectorSimulator(shots=10).run(qc4)

statevector4 = job.result().get_statevector(qc4).data
print(statevector4)

# Print the results.
results = job.result()
counts = results.get_counts(qc4)

display(plot_histogram(counts, title='Quantum Coin Flipping'))
print(counts)
```

### Example 5

Now rotate the qubit by a random angle and apply the $X$ gate to it.


```python
# Z gate allows reflection in the real vectors.
# Once you have a global phase, the roation does not matter any more.
# Global phase has no physical meaning.
# If we start with a super-position state, Z-gate becomes uneffective.


from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import StatevectorSimulator
from qiskit.visualization import plot_histogram
from math import pi
from random import randint

# We define a quantum circuit with one qubit and one bit.
q4 =  QuantumRegister(1) # quantum register with a single qubit
c4 = ClassicalRegister(1) # classical register with a single bit
qc4 = QuantumCircuit(q4,c4) # quantum circuit with quantum and classical registers

rotation_angle = randint(0,100)*pi
qc4.ry(rotation_angle/50,q4[0]) 
qc4.x(q4[0])

# Draw the circuit.
display(qc4.draw(output='mpl'))

qc4.measure(q4[0],c4[0])

# Execute the program 1000 times.
job = StatevectorSimulator(shots=1000).run(qc4)

statevector4 = job.result().get_statevector(qc4).data
print(statevector4)

# Print the results.
results = job.result()
counts = results.get_counts(qc4)

display(plot_histogram(counts, title='Quantum Coin Flipping'))
print(counts)
```

### Example 6

Take the function (depending on an argument $p$) that returns one of the two states $\ket{0}$ with probability $p$ and the state $\ket{+}$ with probability $1-p$. 

```python
from random import randrange

def my_source(p):
    if (randrange(101)/100<p):
        return [1,0]

    # probability 1-p
    else:
        return [2**(-1/2), 2**(-1/2)]
```
