# Tensor products

We had tensor products of vectors and matrices as well. Now will look at how tensor products work together.



### Example 1

Consider vectors
$$
u = \myvector{1 \\ 2 \\ -3}, \quad
v = \myvector{-5 \\ 3 \\ 4},
$$
and matrices
$$
    M = \mymatrix{rrr}{-1 & 0 & 1 \\ -2 & -1 & 2 \\ 1 & 2 & -2} ~~\mbox{and}~~ 
    N = \mymatrix{rrr}{0 & 2 & 1 \\ 3 & -1 & -2 \\ -1 & 1 & 0}.
$$

We compute a new vector $w=u\otimes v$ and new matrix $L=M\otimes N$, then transform $w$ by $L$. We compare this with the tensor product $(Mu)\otimes(Nv)$.


```python
# Vectors u and v
# Vector multiplication
# Tensor product of two vectors.

u = [1,2,-3]  # ---> Vector u.
v = [-5,3,4]  # ---> Vector v.

# We create a third vector, w. And multiply every basis vector of u with those of v. 

w = []
for i in range(len(u)): # each element of v will be replaced
    for j in range(len(v)): # the vector u will come here after multiplying with the entry there
        w.append( u[i] * v[j] )

print("w=uv is:\n",w)
```

```python
# Matrices M and N
# Multiplying matrices or a matrix with a vector is essential. 

# See the notation, We give three vectors into the matrix. Every inner [] rep. a row. 

M = [ 
    [-1,0,1],
    [-2,-1,2],
    [1,2,-2]
]
N = [
    [0,2,1],
    [3,-1,-2],
    [-1,1,0]
]

# L will be a (9x9)-dimensional matrix
# Prepare it as an empty matrix.
# This helps us to easily fill it.
# L is initially an empty list, it will hold the result of the Kronecker product. Since 'M' and 'N' are 3x3 matrices, the result 
# of their Kronecker product will be a 9x9 matrix. 
# A nested loop iterates 9 times to append 9 empty strings  to L, each representing a row of the resulting 9x9 matrix. 
# Inside each of these iterations, another loop iterates 9 times to append 0 to each list, initializing all elements of L to 0. 
# This step prepares L to be filled with the actual results. 

L=[]
for i in range(9): # ----> This line generates 9 rows. 'i' is not important.
    L.append([])  # ---> Now, it becomes a matrix.
    for j in range(9): # -----> Since i =! j --->  This line fills each row with nine zeros.
        L[i].append(0)
        
# Calculation of the Kronecker Product. The calculation involves four nested loops. 

for i in range(3): # row of M
    for j in range(3): # column of M
        for k in range(3): # row of N
            for l in range(3): # column of N
                
# This is how you write the Kronecker product structure. 
# The Kronecker product is not the same as the regular matrix multiplication. 
# Instead of combining rows and columns through dot products, it produces a new, larger matrix by multiplying matrices.
# This operation is particularly useful in tensor product spaces, quantum computing, and systems theory.
                L[3*i+k][3*j+l] = M[i][j] * N[k][l] 

print("L=MN is:\n")   

for i in range(9):
    print(L[i])
```

Consider vectors
$$
u = \myvector{1 \\ 2 \\ -3}, \quad
v = \myvector{-5 \\ 3 \\ 4},
$$
and matrices
$$
    M = \mymatrix{rrr}{-1 & 0 & 1 \\ -2 & -1 & 2 \\ 1 & 2 & -2} ~~\mbox{and}~~ 
    N = \mymatrix{rrr}{0 & 2 & 1 \\ 3 & -1 & -2 \\ -1 & 1 & 0}.
$$

We compute a new vector $w=u\otimes v$ and new matrix $L=M\otimes N$, then transform $w$ by $L$, which means $L \otimes (w)$. We compare this with the tensor product $(Mu)\otimes(Nv)$. We will se soon, a fundamental property of tensor products in linear algebra where transforming a vector by a Kronecker product matrix is equivalent to taking the Kronecker product of the transformations of the vector components separately. Thus, the tensor product is interchangable.


```python
# Compute M*u
# The result vector Mu
Mu = []

# For each row, we do an inner product.
for i in range(3):
    
    # Inner product for one row is initiated
    inner_result = 0 # This variable keeps the summation of the pairwise multiplications.
    
    for j in range(3): # The elements in the i-th row
        inner_result = inner_result + M[i][j] * u[j]
        
    # Inner product for one row is completed. Then, the loop starts over.
    Mu.append(inner_result)
print(" Mu is:\n", Mu)

# Compute N*v
# the result vector Nv
Nv = []

# For each row, we do an inner product
for i in range(3):
    
    # Inner product for one row is initiated.
    inner_result = 0 # This variable keeps the summation of the pairwise multiplications.
    for j in range(3): # The elements in the i-th row
        inner_result = inner_result + N[i][j] * v[j]
        
    # inner product for one row is completed
    Nv.append(inner_result)
print('\n',"Nv is:\n" ,Nv)
```

```python
# Tensor product of Mu and Nv. 

x = []
for i in range(len(Mu)): # Each element of v will be replaced.
    for j in range(len(Nv)): # The vector u will come here after multiplying with the entry there.
        x.append( Mu[i] * Nv[j] )

print("x = NvMu =",x)
```

```python
# Compute L*w.
# The result vector Nv.

# Tensor product is thus interchangable.

y = []

# For each row, we do an inner product.
for i in range(9):
    
    # Inner product for one row is initiated.
    inner_result = 0 # this variable keeps the summation of the pairwise multiplications.
    for j in range(9): # The elements in the i-th row.
        inner_result = inner_result + L[i][j] * w[j]
        
    # Inner product for one row is completed.
    y.append(inner_result)
print("y = Lw =",y)
```

```python
if x == y:
    print('x & y are identical.')
    print('We thus proved that Kronecker product is interchangable.')
```

# Powers of complex numbers

Powers of complex numbers seem to be more complicated to calculate, since we have to raise the sum of two terms onto some power. 

$z=r*e^{i\phi}$ 

However, if we use the polar representation we see that $z^n = r^n e^{i\phi n} = r^n (\cos(n\phi) + i\sin(n\phi))$. Which is Euler formula.

### Example 3
Let's verify this quickly, by taking a complex number $z$ with modulus $|z|=1$.

```python

# Python calls it phase.

from math import sqrt, pi
from cmath import phase

# j equals complex number in Python.
# Our complex number $z$.
z = 0.7 + sqrt(1-0.7*0.7)*1j

print("Absolute value:", abs(z))

# Phase function from cmath lib. calculates the phase angle.
phi=phase(z)

print("Phase phi=",phi/pi,"* ")

# Note, that in python before starting a loop, we first define an empty list, a = []
# We set a variable to zero: a = 0, etc. 

ph=[]
for j in range(10):
    p=phase(z**j)
    ph.append(p)
    print("Phase of z^",j,"=",p/phi,"* phi")

```

# Inner product vs. dot product

The main difference, and the reason for using the inner product instead of the dot product is that a vector inner multiplied by itself will result in a positive (real) number.

### Example 5

Let's see how this works. We take two vectors $\ket{u} = \myvector{1\\ i\\ 1/\sqrt{3}}$ and $\ket{v}=\myvector{3\\ \sqrt{5} \\ 2+i}$, and compute the dot products and the inner product $\braket{u}{v}$

```python
from math import sqrt

# Inner product differs from dot product.

u = [1, 1j, 1/sqrt(3)]
v = [3, sqrt(5), 2+1j]

# Compute inner product k = |

dot=0
inner=0

# To check.
for i in range(len(u)):
    print(u[i].conjugate())

for i in range(len(u)):
    dot = dot + u[i]*v[i]   # ---> Dot product.
    inner = inner + u[i].conjugate()*v[i]   # ---> Inner product. 

print("Dot product:", dot)
print("Inner product:", inner)
               
```

### Example 6

The inner product can be used as a *norm*. If we define $||v||^2 = \braket{v}{v}$, then $w=\frac{v}{||v||}$ will be such that $||w||=1$.

We can code more compactly if we use `numpy` arrays. 

```python
# sum function is used to calculate the total value.

u=np.array(u)
v=np.array(v)

print("Dot product:", sum(u*v))
print("Inner product:", sum(u.conjugate()*v) )

print("'Naive' norm is: ", sum(u*u))
print("Norm of u? ", sum(u.conjugate()*u))
u1 = u/np.sqrt(sum(u.conjugate()*u))
print("Norm ", sum(u1.conjugate()*u1))

print("'Naive' norm is: ", sum(v*v))
v1 = v/np.sqrt(sum(v.conjugate()*v))
print("Norm of v1? ", sum(v1.conjugate()*v1))
print("Norm? (abs) ", abs(sum(v1.conjugate()*v1)))

```

### Example 8

Build a quantum circuit and simulate it, just to see if it works and displays well.

```python
# Import the objects from qiskit

# Instead of bits, qbits. When it is not quantum, it is referred as classical. 

from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit
from qiskit_aer import AerSimulator

# Create a quantum circuit and its register objects.
qreg = QuantumRegister(2) # Quantum register with two quantum bits.
creg = ClassicalRegister(2) # Classical register with two classical bit.
circuit = QuantumCircuit(qreg,creg) # Quantum circuit composed by a quantum register and a classical register.

# Apply a Hadamard gate to the first qubit
circuit.h(qreg[0])

# Set the second qubit to state |1>
circuit.x(qreg[1])

# Apply CNOT(first_qubit,second_qubit)
circuit.cx(qreg[0],qreg[1])

# Measure the both qubits
circuit.measure(qreg,creg)

print("The execution of the cell was completed, and the circuit was created :)")
```

CNOT Gate: 

The CNOT gate, which stands for Controlled-NOT gate, is a two-qubit quantum gate that is central to quantum computing. It operates on a pair of qubits, consisting of a control qubit and a target qubit. The action of the CNOT gate depends on the state of the control qubit:

__If the control qubit is in the state 0, the CNOT gate leaves the state of the target qubit unchanged.__

__If the control qubit is in the state 1, the CNOT gate flips the state of the target qubit (from 0 to 1 or from 1 to 0).__

In essence, the CNOT gate is a conditional gate that performs the NOT operation on the target qubit if and only if the control qubit is in the state 1.


The CNOT gate is crucial in quantum computing for several reasons:

__Entanglement:__ It is capable of generating entangled states from separable states, which is a key resource in quantum computing and quantum information theory.

__Quantum Circuits:__ Together with single-qubit gates, the CNOT gate forms a universal set of gates, meaning any quantum operation can be decomposed into a sequence of CNOT and single-qubit gates.

__Quantum Algorithms:__ It is used in many quantum algorithms and protocols, including quantum teleportation, superdense coding, and as part of quantum error correction schemes.

The versatility and fundamental nature of the CNOT gate make it one of the most widely used gates in quantum computing experiments and theory.

```python
from qiskit import QuantumCircuit

# Create a Quantum Circuit.
circuit = QuantumCircuit(2)  # For example, a 2-qubit circuit.
circuit.h(0)  # Apply Hadamard gate to the first qubit.
circuit.cx(0, 1)  # Apply CNOT gate.

# Draw the circuit.
circuit.draw(output='mpl')
```
